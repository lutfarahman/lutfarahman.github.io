```python
raw = 'HSSHHHSSHHSHSSHsHHSSHSHSHSSSHHS' #XZYZHSHSH
output = ''

for i in range(len(raw)):   
    print(i)
    if len(output) == 0:
        output += raw[i]  
        
    elif raw[i] == 'H':
        if output[-1] == 'H':     #If input string is H & last element of output is also H.
            output = output[:-1]
        elif len(output) == 1:
                pass    
        elif output[-1] == 'Z' and output[-2] == 'H':   #HZH = X, ZH = W and HW = X
            output = output[:-1] + 'W'
            if len(output) == 1:
                pass
            
            elif output[-1] == 'W' and output[-2] == 'H':
                output = output[:-2] + 'X' 
            elif output[-1] == 'W' and raw[i] == 'H': #ZH = W, ZHH = WH=Z
                output = output[:-1] + 'Z'
            
        else:
            output += raw[i]
            
    elif raw[i] == 'S':
        if output[-1] == 'S':    #SS = Z
            output = output[:-1] + 'Z' 
            if len(output) == 1:
                pass
            elif output[-2] == 'Z': #ZZ = I
                output = output[:-2]
        else:
            output += raw[i]
            
    elif raw[i] == "s": #S-dagger
        if output[-1] == 'X': #SXs = Y
            output = output[:-1] + 'a'
            if output[-1] == 'a' and output[-2] == 'S':
                output = output[:-2] + 'Y' 
            
    else:
        output += raw[i] 
                
    print(output, raw[i]) 
```
